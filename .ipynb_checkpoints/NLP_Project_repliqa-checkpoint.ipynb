# Pipeline for automated translation
__________________________________________________________________________



This Jupyter notebook implements the data verification pipeline for translating an English dataset to Arabic. We use **NLLB** (No Language Left Behind) models for translation instead of GPT-4o-mini.


**The pipeline includes:**

1. Forward translation (EN to AR)

2.   Semantic similarity check using LaBSE embeddings and cosine similarity (threshold: â‰¥80%)
3.   Reverse translation (AR to EN)

4.    Quality metrics evaluation: BLEU (>60%), METEOR (>80%), ROUGE Precision/Recall/F1 (>80%)

 5. Filtering: Only rows that pass all checks are included in the output dataset.


**Prerequisites:**


*   Install required libraries: Run the first code cell to install them.
*   Place your input CSV file (e.g., **input_dataset**.csv) in the same directory.
*  The output will be saved as **translated_dataset.csv** with columns: **id, en-text, ar-text, label**.

```python
# --- Install once per runtime ---

import os, time, math, json, pandas as pd
from tqdm import tqdm
import torch
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, pipeline

# ====== CONFIG ======
INPUT_CSV  = "/content/repliqa_0_en.csv"
OUTPUT_CSV = "/content/repliqa_0_ar.csv"
OUTPUT_CSV_DRIVE = "/content/drive/MyDrive/NLP/repliqa_0_ar.csv"

TARGET_TOPIC = "local technology and innovation"  # case-insensitive contains match
BATCH_SIZE = 8                                    # you can raise if GPU memory allows
USE_MODEL = "nllb"                                # "nllb" (better) or "helsinki" (lighter)

SOURCE_COLS = ["document_topic", "document_extracted", "question", "answer", "long_answer"]
TARGET_COLS = ["document_topic_ar", "document_extracted_ar", "question_ar", "answer_ar", "long_answer_ar"]

# ====== Load & filter ======
df = pd.read_csv(INPUT_CSV, dtype=str, encoding="utf-8-sig")
print("Loaded:", INPUT_CSV, "Rows:", len(df))

df_topic = df[df["document_topic"].str.lower().str.contains(TARGET_TOPIC.lower(), na=False)].copy()
print(f"ðŸŽ¯ Selected topic rows containing '{TARGET_TOPIC}' â€” {len(df_topic)} rows")
if len(df_topic) == 0:
    raise ValueError(f"No rows found for topic: {TARGET_TOPIC}")

# Ensure cols exist
for c in SOURCE_COLS:
    if c not in df_topic.columns:
        df_topic[c] = ""
for c in TARGET_COLS:
    if c not in df_topic.columns:
        df_topic[c] = ""

# ====== Build local translation pipeline ======
device_id = 0 if torch.cuda.is_available() else -1
print("CUDA available:", torch.cuda.is_available())

if USE_MODEL.lower() == "nllb":
    model_id = "facebook/nllb-200-distilled-600M"  # good quality
    tok = AutoTokenizer.from_pretrained(model_id)
    mdl = AutoModelForSeq2SeqLM.from_pretrained(model_id)
    translator = pipeline(
        "translation",
        model=mdl, tokenizer=tok,
        src_lang="eng_Latn", tgt_lang="arb_Arab",
        device=device_id
    )
else:
    model_id = "Helsinki-NLP/opus-mt-en-ar"        # lighter/faster
    tok = AutoTokenizer.from_pretrained(model_id)
    mdl = AutoModelForSeq2SeqLM.from_pretrained(model_id)
    translator = pipeline(
        "translation",
        model=mdl, tokenizer=tok,
        device=device_id
    )

# ====== Small cache to avoid re-translating repeats ======
TEXT_CACHE = {}

# Pre-fill cache from existing AR values (resume-safe)
for src_col, tgt_col in zip(SOURCE_COLS, TARGET_COLS):
    both = df_topic[[src_col, tgt_col]].dropna()
    for en, ar in both.values:
        en_s = str(en).strip()
        ar_s = str(ar).strip()
        if en_s and ar_s:
            TEXT_CACHE[en_s] = ar_s

def translate_many(text_list):
    """Translate a list with caching and batching."""
    out = []
    batch = []
    idxs = []
    # Collect uncached
    for i, t in enumerate(text_list):
        t = "" if pd.isna(t) else str(t)
        tt = t.strip()
        if not tt:
            out.append("")  # keep alignment
            continue
        if tt in TEXT_CACHE:
            out.append(TEXT_CACHE[tt])
        else:
            out.append(None)  # placeholder to fill later
            batch.append(tt)
            idxs.append(i)

    # Run model for uncached items in mini-batches
    for j in range(0, len(batch), BATCH_SIZE):
        chunk = batch[j:j+BATCH_SIZE]
        res = translator(chunk, max_length=1024, truncation=True)
        trans = [r["translation_text"] for r in res]
        for k, arabic in enumerate(trans):
            t_src = chunk[k]
            TEXT_CACHE[t_src] = arabic

    # Fill placeholders from cache
    for i, val in enumerate(out):
        if val is None:
            src_text = text_list[i].strip()
            out[i] = TEXT_CACHE.get(src_text, "")

    return out

# ====== Translate column by column (only empty targets) ======
for src_col, tgt_col in tqdm(zip(SOURCE_COLS, TARGET_COLS), total=len(SOURCE_COLS), desc="Translating columns"):
    needs_mask = df_topic[tgt_col].isna() | (df_topic[tgt_col].astype(str).str.strip() == "")
    if needs_mask.any():
        src_texts = df_topic.loc[needs_mask, src_col].astype(str).tolist()
        ar_texts = translate_many(src_texts)
        df_topic.loc[needs_mask, tgt_col] = ar_texts

# ====== Save output (cache + Drive) ======
os.makedirs(os.path.dirname(OUTPUT_CSV_DRIVE), exist_ok=True)
df_topic.to_csv(OUTPUT_CSV, index=False, encoding="utf-8-sig")
df_topic.to_csv(OUTPUT_CSV_DRIVE, index=False, encoding="utf-8-sig")
print(f"âœ… Saved {OUTPUT_CSV} and {OUTPUT_CSV_DRIVE}  ({len(df_topic)} rows)")

print("Columns:", list(df_topic.columns))
```
